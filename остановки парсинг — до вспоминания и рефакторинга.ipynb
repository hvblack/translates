```python
# определяем чистый фрейм - без логотипов и инструментов
class r:
    top=128
    bottom=976
    left=432
    right=1824
#up_left=[432,128] # правый нижний угол прокладчика маршрута
#down_right=[1856,976] # этажность (выскакивает слева от линейки и масштаба) и верх логотипа 2GIS + чтобы итоговый фрейм кратен 8
figure_size=16

def movement(r,direction):
    if direction=='left':
        pyautogui.moveTo(r.left,r.top,duration=0.2)
        #sleep(1)
        pyautogui.mouseDown(button='left')
        sleep(0.5)
        pyautogui.move((r.right-r.left-16), 0,duration=0.2)
        sleep(0.5)
        pyautogui.mouseUp(button='left')
        pyautogui.moveTo(1,1)
    elif direction=='right':
        pyautogui.moveTo(r.right,r.top,duration=0.2)
        #sleep(1)
        pyautogui.mouseDown(button='left')
        sleep(0.5)
        pyautogui.move(-(r.right-r.left-16), 0,duration=0.2)
        sleep(0.5)
        pyautogui.mouseUp(button='left')
        pyautogui.moveTo(1,1)
    elif direction=='top':
        pyautogui.moveTo(r.left,r.top,duration=0.2)
        #sleep(1)
        pyautogui.mouseDown(button='left')
        sleep(0.5)
        pyautogui.move(0,(r.bottom-r.top-16),duration=0.2)
        sleep(0.5)
        pyautogui.mouseUp(button='left')
        pyautogui.moveTo(1,1)
    elif direction=='bottom':
        pyautogui.moveTo(r.left,r.bottom,duration=0.2)
        #sleep(1)
        pyautogui.mouseDown(button='left')
        sleep(0.5)
        pyautogui.move(0,-(r.bottom-r.top-16),duration=0.2)
        sleep(0.5)
        pyautogui.mouseUp(button='left')
        pyautogui.moveTo(1,1)
        

import webbrowser
webbrowser.register('my_chrome', None, webbrowser.BackgroundBrowser('C:\Program Files (x86)\Google\Chrome\Application\chrome.exe'))


import pyautogui
import pyperclip

from time import sleep
from PIL import Image, ImageGrab

import numpy as np
import pandas as pd

import json
from my_functions import *

# для распараллеливания
from joblib import Parallel, delayed

from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt

def to_vector(pix):
    #selection=pix[:,:,2]>200
    #temp_pix[~selection]
    #pix=pix[selection]
    temp_pix=pix.copy()
    temp_pix[temp_pix[:,:,2]<180]=255#np.array([255,255,255])
    etalon=temp_pix.mean(axis=1).mean(axis=0)
    if pix.shape[0]==0:
        return np.zeros((6))
    #etalon=pix.mean(axis=0)
    if etalon.mean()>0:
        etalon=etalon/etalon.mean()
    std=temp_pix.mean(axis=1).std(axis=0)
    #std=pix.std(axis=0)
    if std.mean()>0:
        std=std/std.mean()
    etalon=np.append(etalon,std)
    
    return etalon

def create_vector(pix):
    pix=pix[1:15,1:15]
    etalon=to_vector(pix)
    #etalon_svod.append(etalon)
    
    temp=to_vector(pix[:7,:7])
    etalon_full=np.append(etalon,temp)
    temp=to_vector(pix[:7,7:])
    etalon_full=np.append(etalon_full,temp)
    temp=to_vector(pix[7:,:7])
    etalon_full=np.append(etalon_full,temp)
    temp=to_vector(pix[7:,7:])
    etalon_full=np.append(etalon_full,temp)
    
    #etalon_svod_full.append(etalon)

    return etalon,etalon_full
    
# переводим значки в векторы
etalon_svod=[]
etalon_svod_full=[]
for filename in ['2gis ост1 19.png','2gis ост2 19.png','2gis станция 19.png']:
    image = Image.open(filename)
    pix = np.array(image)
    pix=pix[:,:,:3]
    etalon,etalon_full=create_vector(pix)
    
    etalon_svod.append(etalon)
    etalon_svod_full.append(etalon_full)

branches=df_build('mart_com','branches')


from scipy.spatial.distance import euclidean
def evklid(a,b):
    #return np.sqrt(sum( (x1 - x2)**2 for x1, x2 in zip(a, b)))
    #return np.sqrt(sum((a-b)**2))
    return euclidean(a,b)



def save_source():
    #pyautogui.moveTo(1920/2,1080/2)
    #pyautogui.click()
    x=datetime.datetime.now()
    #sleep(2) - до этого давно не работаем с брарузером
    pyautogui.keyDown('ctrl')
    pyautogui.press('u')
    pyautogui.keyUp('ctrl')
    
    sleep(2) # загрузка исходного кода
    
    #pyautogui.click() # рискуем попасть на ссылку в исходном коде
    #sleep(2)
    pyautogui.keyDown('ctrl')
    pyautogui.press('a')
    pyautogui.keyUp('ctrl')
    sleep(0.5) # выделение
    
    pyautogui.keyDown('ctrl')
    pyautogui.press('c')
    pyautogui.keyUp('ctrl')
    
    sleep(0.5) # копирование
    
    pyautogui.keyDown('ctrl')
    pyautogui.press('w')
    pyautogui.keyUp('ctrl')
    y=datetime.datetime.now()
    print('save_source',str(y-x))
    x = pyperclip.paste()
    return x

def parse_source(source):
    cnt=0
    begin=source.find('geometry')
    for i in range(begin+10,len(source)):
        if source[i]=='{':
            cnt+=1
        elif source[i]=='}':
            cnt-=1
        if cnt==0:
            geometry=source[begin+10:i+1].replace('=\n','')
            break
    
    cnt=0
    begin=source.find('routes')
    if begin>=0:
        for i in range(begin+8,len(source)):
            if source[i]=='[':
                cnt+=1
            elif source[i]==']':
                cnt-=1
            if cnt==0:
                routes=source[begin+8:i+1].replace('=\n','')
                break
    else:
        routes='[]'
            
    cnt=0
    begin=source.find('startURL')
    for i in range(begin+10,len(source)):
        if source[i]==',':
            startURL=source[begin+10:i].replace('=\n','')
            break
    
    geometry=json.loads(geometry)
    geometry['routes']=json.loads(routes)
    #print(startURL)
    geometry['startURL']=json.loads(startURL)
    #return [json.loads(geometry),json.loads(routes)]
    return geometry

def create_metrics(temp_pix,etalon_svod):
    #act=temp.mean(axis=1).mean(axis=0)
    #act=act/act.mean()
    #std=temp.mean(axis=1).std(axis=0)
    #if std.mean()>0:
    #    std=std/std.mean()
    #act=np.append(act,std)
    
    # убираем углы. там лишнее
    act,act_full=create_vector(temp_pix)
    ev_all =[evklid(act,et) for et in etalon_svod]
    #ev2=evklid(act,etalon_svod[1])
    
    return min(ev_all)

def create_metrics_full(pix,x_ono,y_ono,figure_size,etalon_svod_full):
    temp_pix=pix[int(y_ono):int(y_ono+figure_size),int(x_ono):int(x_ono+figure_size)]
    # убираем углы. там лишнее
    act,act_full=create_vector(temp_pix)
    ev_all =[evklid(act_full,et) for et in etalon_svod_full]
    #print(act)
    #print(act_full)
    #print(ev_all)
    #ev2=evklid(act,etalon_svod[1])
    return min(ev_all)

   
    
def find_stations(img,r,step=8,figure_size=16):
    #print('begin find station',datetime.datetime.now())
    pix = np.array(img)[:,:,:3]
    
    cnt=0
    coordinates=[]
    dist=np.zeros((int((r.bottom-r.top)/step),int((r.right-r.left)/step)))
    #plt.figure()
    for y in range(r.top,r.bottom,step):
        for x in range(r.left,r.right,step):
            #print('upper squ',datetime.datetime.now())
            temp=pix[y:(y+figure_size),x:(x+figure_size)]
            ev1=create_metrics(temp,etalon_svod)

            #print(int((x-x1)/step),int((y-y1)/step),ev1,act,temp)
            dist[int((y-r.top)/step),int((x-r.left)/step)]=ev1
            if ev1<0.2:
                
                cnt+=1
                print()
                print(cnt,x,y,ev1,(x-r.left)/step,(y-r.top)/step)
                min_dist=1000
                #print('lower squ begin',datetime.datetime.now())
                #for xx in range(x-figure_size,x+figure_size,4):
                #    for yy in range(y-figure_size,y+figure_size,4):
                        #temp=pix[yy:(yy+figure_size),xx:(xx+figure_size)]
                        #ev1=create_metrics(temp,etalon_svod)
                #        ev1=lower_check(pix,xx,yy)[0]
                #        if ev1<min_dist:
                #            min_dist=ev1
                #            x_ono=xx
                #            y_ono=yy
                        #print(cnt,x,y,xx,yy,ev1)
                #print('lower squ middle',datetime.datetime.now())
                all_dist=Parallel(n_jobs=12)(delayed(lower_check)(pix,xx,yy)\
                          for xx in range(x-figure_size,x+figure_size,4)\
                            for yy in range(y-figure_size,y+figure_size,4))
                #print('lower squ middle after parall',datetime.datetime.now())
                df=pd.DataFrame(all_dist)
                df.columns=['dist','x','y']
                df=df.sort_values('dist',ascending=True).head(1)
                x_ono=int(df.x.values)
                y_ono=int(df.y.values)
                #print(x_ono,y_ono)
                #print('lower squ end',datetime.datetime.now())
                #dist_middle=create_dist_middle(pix,x_ono,y_ono,figure_size)
                #print(dist_middle)
                #if dist_middle<15:
                #full_equality=create_metrics_full(pix[y_ono:int(y_ono+figure_size),x_ono:int(x_ono+figure_size)],etalon_svod_full)
                full_equality=create_metrics_full(pix,x_ono,y_ono,figure_size,etalon_svod_full)
                if full_equality<0.1:
                    #plt.figure()
                    #plt.imshow(pix[y_ono-32:(y_ono+figure_size+32),x_ono-32:(x_ono+figure_size+32)])
                    #plt.figure()
                    #plt.imshow(pix[y_ono:(y_ono+figure_size),x_ono:(x_ono+figure_size)])
                    #print(x_ono,y_ono,create_metrics_full(pix[y_ono:int(y_ono+figure_size),x_ono:int(x_ono+figure_size)],etalon_svod_full))
                    coordinates.append([x_ono+figure_size/2,y_ono+figure_size/2])

                # зануляем найденный регион (даже есди не остановка, чтобы снова не появлялась (т.к. может попасть в другой квадрат)
                #
                pix[y_ono:(y_ono+figure_size),x_ono:(x_ono+figure_size)]=0

                
    #print('begin stops',datetime.datetime.now())
                
                
    
    stops=[]
    for c in coordinates:
        pyautogui.moveTo(c[0],c[1],duration=1)
        pyautogui.click()
        sleep(1)

        source=save_source()
        if '<!-- -->Станция' in source or '<!-- -->Остановка наземного транспорта' in source:
            print(c,'Остановка')
            temp=parse_source(source)
            stops.append(temp)
            sleep(1)
            
    img.save('stop screenshots/'+datetime.datetime.now(tz).strftime("%Y-%m-%d %H-%M-%S")+' '+ str(len(stops))+
              ".png", "PNG", quality=80, optimize=True, progressive=True)
    
    return dist,coordinates,stops,all_dist


def upper_check(pix,x,y):
    temp=pix[y:(y+figure_size),x:(x+figure_size)]
    ev1=create_metrics(temp,etalon_svod)
    return [ev1,x,y]

def lower_check(pix,xx,yy):
    temp=pix[yy:(yy+figure_size),xx:(xx+figure_size)]
    ev1=create_metrics(temp,etalon_svod)
    return [ev1,xx,yy]

def lower_check2(pix,row):
    temp=pix[int(row.yy):int(row.yy+figure_size),int(row.xx):int(row.xx+figure_size)]
    ev1=create_metrics(temp,etalon_svod)
    return [ev1,row.x1,row.x2,row.y1,row.y2,row.xx,row.yy]

def lower_check3(pix,xx,yy):
    figure_size=16
    temp=pix[int(yy):int(yy+figure_size),int(xx):int(xx+figure_size)]
    ev1=create_metrics(temp,etalon_svod)
    return ev1

def lower_check_main(df,pix,step):
    x=datetime.datetime.now()
    #print(f'begin lower {step}',datetime.datetime.now())
    try:
        df=df.drop(columns=['xx','yy'])
    except:
        pass
    #print(df)
    xdigits=pd.DataFrame(range(int(df.x1.min()),int(df.x2.max()),step),columns=['xx'])
    ydigits=pd.DataFrame(range(int(df.y1.min()),int(df.y2.max()),step),columns=['yy'])
    #print(f'lower cross',datetime.datetime.now())
    df=df.merge(xdigits,how='cross')

    df=df.query('xx>x1 and xx<x2')

    df=df.merge(ydigits,how='cross')
    lower_df=df.query('yy>y1 and yy<y2')
    #print(f'lower dist old',datetime.datetime.now())
    result=Parallel(n_jobs=8)(delayed(lower_check2)(pix,row)\
                          for i,row in lower_df.iterrows())
    lower_df=pd.DataFrame(result)
    lower_df.columns=['dist','x1','x2','y1','y2','xx','yy']
    
    #print(f'lower dist',datetime.datetime.now())
    #lower_df['dist']=lower_df.apply(lambda df:lower_check3(pix,df.xx,df.yy),axis=1)
    #print(f'lower ra',datetime.datetime.now())
    
    lower_df['ra']=lower_df.groupby(['x1','x2','y1','y2'])\
                        ['dist'].rank('dense',ascending=True)
    #to_csv(lower_df,f'lower_df{step} full')
    lower_df=lower_df.query('ra==1')
    #print(f'lower new xy',datetime.datetime.now())
    #to_csv(lower_df)
    #to_csv(lower_df,f'lower_df{step} before')
    lower_df.loc[:,'x1'] = lower_df.apply(lambda df:max(df.x1,df.xx-2*step), axis=1)
    lower_df.loc[:,'y1'] = lower_df.apply(lambda df:max(df.y1,df.yy-2*step), axis=1)
    lower_df.loc[:,'x2'] = lower_df.apply(lambda df:min(df.x2,df.xx+2*step), axis=1)
    lower_df.loc[:,'y2'] = lower_df.apply(lambda df:min(df.y2,df.yy+2*step), axis=1)
    #to_csv(lower_df,f'lower_df{step} after')
    
    #to_csv(lower_df,f'lower_df{step}')
    y=datetime.datetime.now()
    print(f'lower {step}',str(y-x))
    return lower_df


def create_bad_duplicates(lower_df):
    ld=lower_df.reset_index().merge(lower_df.reset_index(),how='cross')
    ld.loc[:,'dist_stops']=abs(ld.xx_x-ld.xx_y)+abs(ld.yy_x-ld.yy_y)
    ld=ld.query('not (xx_x==xx_y and yy_x==yy_y)')
    #ld=ld.sort_values(['dist_stops','index_y'],ascending=[True,True]).reset_index()
    #ld['for_sort']=ld.index

    #ld['ra']=ld.groupby('index_x')['for_sort'].rank('dense',ascending=[True,True])
    ld=ld.query('dist_stops<6 and full_equality_x<full_equality_y')[['xx_y','yy_y']]
    ld.columns=['xx','yy']
    ld['bad_duplicate']=1
    ld=ld.drop_duplicates()
    return ld


def find_stations_new(img,r,step=16,figure_size=16,upper_accuracy=0.2):
    print('begin find station',datetime.datetime.now())
    pix = np.array(img)[:,:,:3]
    
    cnt=0
    stops=[]
    #dist=np.zeros((int((r.bottom-r.top)/step),int((r.right-r.left)/step)))
    #plt.figure()
    x=datetime.datetime.now()
    result=Parallel(n_jobs=8)(delayed(upper_check)(pix,x,y)\
                          for y in range(r.top,r.bottom,step)\
                            for x in range(r.left,r.right,step))
    
    upper_df=pd.DataFrame(result)
    upper_df.columns=['dist','x','y']
    to_csv(upper_df,'upper_full')
    upper_df=upper_df.query(f'dist<{upper_accuracy}')
    if upper_df.empty:
        img.save('stop screenshots/'+datetime.datetime.now(tz).strftime("%Y-%m-%d %H-%M-%S")+' '+ str(len(stops))+
              ".png", "PNG", quality=80, optimize=True, progressive=True)
        return [],[],[],[]
    # раньше было 16. но если пропустим в одном, то найдём в другом квадранте
    upper_df.loc[:,'x1']=upper_df.x-12
    upper_df.loc[:,'x2']=upper_df.x+12
    upper_df.loc[:,'y1']=upper_df.y-12
    upper_df.loc[:,'y2']=upper_df.y+12

    y=datetime.datetime.now()
    print('upper',str(y-x))
    print(upper_df.shape[0])
    if upper_df.empty:
        img.save('stop screenshots/'+datetime.datetime.now(tz).strftime("%Y-%m-%d %H-%M-%S")+' '+ str(len(stops))+
              ".png", "PNG", quality=80, optimize=True, progressive=True)
        return [],[],[],[]
    
    lower_df=lower_check_main(upper_df,pix,8)
    lower_df=lower_df.query('dist<0.2')
    print(lower_df.shape[0])
    if lower_df.empty:
        img.save('stop screenshots/'+datetime.datetime.now(tz).strftime("%Y-%m-%d %H-%M-%S")+' '+ str(len(stops))+
              ".png", "PNG", quality=80, optimize=True, progressive=True)
        return [],[],[],[]
    
    lower_df=lower_check_main(lower_df,pix,4)
    lower_df=lower_df.query('dist<0.1')
    print(lower_df.shape[0])
    if lower_df.empty:
        img.save('stop screenshots/'+datetime.datetime.now(tz).strftime("%Y-%m-%d %H-%M-%S")+' '+ str(len(stops))+
              ".png", "PNG", quality=80, optimize=True, progressive=True)
        return [],[],[],[]
                            
    lower_df=lower_check_main(lower_df,pix,2)
    lower_df=lower_df.query('dist<0.05')
    print(lower_df.shape[0])
    if lower_df.empty:
        img.save('stop screenshots/'+datetime.datetime.now(tz).strftime("%Y-%m-%d %H-%M-%S")+' '+ str(len(stops))+
              ".png", "PNG", quality=80, optimize=True, progressive=True)
        return [],[],[],[]
                            
    lower_df=lower_check_main(lower_df,pix,1)
    
    x=datetime.datetime.now()
    lower_df['full_equality']=lower_df.apply(lambda df:create_metrics_full(pix,df.xx,df.yy,figure_size,etalon_svod_full),axis=1)
    #print(lower_df)
    lower_df=lower_df.query('full_equality<0.1')
    
    ld=create_bad_duplicates(lower_df)
    lower_df=lower_df.merge(ld,how='left',on=['xx','yy'])
    lower_df=lower_df[lower_df.bad_duplicate.isna()]
    
    coordinates=lower_df[['xx','yy']].drop_duplicates()
    y=datetime.datetime.now()
    print('full_equality',str(y-x))
    
    for i,row in coordinates.iterrows():
        #
        pyautogui.moveTo(row.xx+8,row.yy+8,duration=0.3)
        sleep(0.5)
        pyautogui.click()
        sleep(1)

        source=save_source()
        
        if '<!-- -->Станция' in source or '<!-- -->Остановка наземного транспорта' in source:
            print(row,'Остановка')
            #print(c,'Остановка')
            #print(parse_source(source))
            stops.append(parse_source(source))
            sleep(0.2) # не помню зачем, но пусть будет
            
    img.save('stop screenshots/'+datetime.datetime.now(tz).strftime("%Y-%m-%d %H-%M-%S")+' '+ str(len(stops))+
              ".png", "PNG", quality=80, optimize=True, progressive=True)
    print('end find station',datetime.datetime.now())
    return coordinates,stops,lower_df,upper_df


def main(stops_svod):
    sleep(2) # на прогрузку
    print('begin', datetime.datetime.now())
    img=ImageGrab.grab((0,0,1920,1080))
    sleep(1) # на скриншот
    #dist,coordinates,stops,all_dist=find_stations(img,r,step=8)
    coordinates,stops,lower_df,upper_df=find_stations_new(img,r,step=16)
    if len(stops)>0:
        stops_svod['Data'].extend(stops)
    print('end', datetime.datetime.now())   
    return stops_svod
```
