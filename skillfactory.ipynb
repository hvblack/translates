```python
raw_list = [2, 8, 10, 23, 64, 49, 11, 52, 71, 14]
x_min=min(raw_list)
x_max=max(raw_list)
my_list=[]
for i in raw_list:
    if i % 2==0:
        my_list.append(i*x_min)
    else:
        my_list.append(i*x_max)
result=max(my_list)
```

```python
# группировка по расе и образованию
students = {}

f = open('StudentsPerformance.csv')

for line in f:
    info = line.split(',')
    if info[0] == '"gender"':
        continue
    else:
        ethnicity = info[1][1:-1]
        parents = info[2][1:-1]
        if ethnicity in students:
            if parents in students[ethnicity]:
                students[ethnicity][parents] += 1
            else:
                students[ethnicity][parents] = 1
        else:
            students[ethnicity] = {}
            students[ethnicity][parents] = 1  
```

```python
from copy import copy

def remove_dups(values):
    values = copy(values)
    i=len(values)-1
    while i >=0:
        if values[i] in values[:i]:
            values.pop(i)
        i-=1
    return values
```

```python
remove_dups([1, 12, 4, 1, 4, 8])
```

```python
from collections import defaultdict

def group_values(db, value_key, group_key, step):
    grouped = defaultdict(list) 
    for item in db:
        grouped[step*(item[group_key] // step)].append(item[value_key])
    return grouped
```

```python
group_values(user_db, "salary", "age", 10)
```

```python
class DepartmentReport():  
    
    def __init__(self,company_name):  
        self.revenues =[] #список, где мы храним значения выручки отделов
        self.company_name=company_name
     
    def average_revenue(self):
        if len(self.revenues)==0:
            mm=0 
        else:
            mm=sum(self.revenues)/len(self.revenues)
        return "Average department revenue for {}: {}".format(self.company_name,round(mm))
    
    def add_revenue(self,amount):
        self.revenues.append(amount)
```

```python
qq=DepartmentReport('Danon')
```

```python
qq.add_revenue(300)
qq.add_revenue(500)
qq.add_revenue(700)
qq.revenues
```

```python
qq.average_revenue()
```

```python
class User():  
    
    def __init__(self,email, password, balance):  
        self.email=email
        self.password=password
        self.balance=balance
     
    def login(self,email,password):
        if self.email==email and self.password==password:
            return True
        else:
            return False
    
    def update_balance(self,amount):
        self.balance+=amount
```

```python
user = User("gosha@roskino.org", "qwerty", 20_000)
print(user.login("gosha@roskino.org", "qwerty123"))
# => False
print(user.login("gosha@roskino.org", "qwerty"))
# => True
user.update_balance(200)
user.update_balance(-500)
print(user.balance)
# => 19700
```

```python
class IntDataFrame():  
    def __init__(self, spis):  
        self.spis = [int(x) for x in spis]
          
    def count(self):  
        return len([x for x in self.spis if x>0])
                  
    def unique(self):
        spis=self.spis
        return len ([spis[i] for i in range(len(spis)) if spis[i] not in spis[(i+1):]])
```

```python
df = IntDataFrame([4.7, 4, 3, 0, 2.4, 0.3, 4])
df.spis
```

```python
df.count()
df.unique()
```

```python
class OwnLogger():  
    def __init__(self):  
        self.spis = {"info":None, "warning":None, "error":None, "all":None}
    
    def log(self,message, level):
        self.spis[level]=message
        self.spis['all']=message
    
    def show_last(self,level='all'):
        return self.spis[level]
```

```python
logger = OwnLogger()
logger.log("System started", "info")
print(logger.show_last("error"))
# => None
# Некоторые интерпретаторы Python могут не выводить None, тогда в этой проверке у вас будет пустая строка
logger.log("Connection instable", "warning")
logger.log("Connection lost", "error")

print(logger.show_last())
# => Connection lost
print(logger.show_last("info"))
# => System started
```
